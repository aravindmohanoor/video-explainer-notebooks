```python
import re

for one_line in open('words'):
    m = re.search('a.*e.*p.*m', one_line)
    if m:
        print(one_line, end='')
```

```python
import re

ro = re.compile('a.*e.*p.*m')

for one_line in open('words'):
    m = ro.search(one_line)
    if m:
        print(one_line, end='')
```
